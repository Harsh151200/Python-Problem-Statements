```python
from nsepy import get_history
from datetime import date

data = get_history(symbol = 'INFY', start = date(2020,1,15), end = date(2021,1,15))
data[['Close']].plot()
```

```python
infy = get_history(symbol = 'INFY', 
                    start = date(2020,1,15), 
                    end = date(2021,1,15))
infy
```

```python
# Stock futures (Similarly for index futures, set index = True)
stock_fut = get_history(symbol="INFY",
                            start = date(2021,1,16), 
                            end = date(2021,1,20),
                            futures=True,
                            expiry_date=date(2021,1,21))
stock_fut
```
